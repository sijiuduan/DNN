# 神经网络 线性回归

##  线性回归的的问题，用得着神经网络来处理吗？
最开始接触神经网络时，我有个疑问，线性回归的的问题，用得着神经网络来处理吗？（当然是有用，当时没想明白）  
我写的第一个神经网络的代码，拿了一个简单的线性回归问题去跑。  
当时是图方便测试，和有助于我理解神经网络工作原理。  
和直接的线性回归做比较，当时[4,4,4,1]的线性神经网络，效率会慢（虽然迭代次数少，但是神经元多，总计算量实际上略大）并没有优势。  

以前看过 SLAM 方面一些资料，现在用神经网络的概念回想，感觉轻松多了。用特定结构的神经网络去处理SLAM问题（例如替换RANSAC过程），让SLAM变得简单了。我想这可以算是神经网络在线性回归问题上的应用吧。

我后来发现，有人做过类似的研究《Spatial Transformer Networks》，甚至提出了“空间变换神经网络”的概念。  
感兴趣的可以去参考：https://www.cnblogs.com/neopenx/p/4851806.html

----

抛开形而上的理解，下面，让我们来看看如何搭建神经网络，处理一个最简单的线性回归问题，

```python
import sys;
sys.path.append("../")
from jnn.plot import *
from jnn.nn import *
from jnn.dataFactory import *
from jnn.activFun import eaActivFun
from jnn.costFun import eaCostFun
from jnn.ea import Ea
```

## 生成一个线性模型的测试数据
  
> $y = 2\times{x_1} - 3.4\times{x_2} + 4.2 + noise$

```python
true_w = [2, -3.4]
true_b = [4.2]
eaData = eaDataLinear(true_w,true_b, m=1000)
```

## 设计一个线性神经网络结构生成器

> 可以已指定网络层数 L, 和每层的神经元个数 n.  
这里用到了 超级字典 Ea

```python
def eaNetLinear(eaData,L,n):
    net = Ea()
    net.L = L
    net.n[0] = eaData.X.shape[0]  # nx
    for l in range(1,L):
        net.n[l]=n
        net.activFun[l] = eaActivFun.linear
    net.n[L] = 1
    net.activFun[L] = eaActivFun.linear

    net.costFun = eaCostFun.L2
    return net
```

## 根据数据 构造4层线性神经网络  
> 用到了输入数据矢量 $\vec{x}$ 的纬度 $n_x$  作为网络的 $n[0]$ 的值

```python
net = eaNetLinear(eaData,L=4,n=4)
Ea.show(net)
```

## 随机初始化参数W, b
> 我们要理解，为什么要随机初始化W,b? 相同初始化的(W,b)，计算梯度时都是相同的。训练后神经元只能相同，和单神经元没区别。

```python
net = nnInitWb(net)
```

## 开始训练 神经网络  
 nnFit(...)
> 默认10000次迭代，可以修改。  
> 加入了自动梯度截断功能。（这个以后再细讲）

训练完成，我们得到了 (w,b) 和每次迭代的损失函数计算值costs。

```python
net = nnFit(eaData, net, learn_rate=0.02)

Ea.show(net.W,3)
Ea.show(net.b,3)
Ea.show(net.costs,3)
```

## 绘制损失函数收敛过程

4层，每层4个神经元的网络，经过20多次迭代基本上就收敛了。  
再看看我们的梯度截断，工作的也不错

```python
plotCost("Linear Costs", net.costs,netInfo=eaNetInfo(net))
```

## 当L=1是，相当于只有输出层单神经元的网络，相当于传统线性回归
单神经元时，每次迭代计算量，是上边13个神经元的网络的 $\frac{1}{13}$   
收敛速度随迭代次数增加，越来越慢。  
在相同的梯度截断控制参数作用下，最终的收敛值为8.25，远不如之前4层神经网络的0.12   

----

怪不得，之前看到过，线性回归需要伴随迭代次数，动态调整学习率了。有时间的时候，我也给我的 jnn 加个学习率自动优化功能。  


```python
net = eaNetLinear(eaData,L=1,n=4)
net = nnInitWb(net)
net = nnFit(eaData, net, learn_rate=0.02)
plotCost("Linear Costs", net.costs,netInfo=eaNetInfo(net))
```
